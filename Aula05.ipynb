```python
from google.colab import userdata
chat_api = userdata.get('api_chat26')
```

```python
import gradio as gr
from groq import Groq
```

```python
  client = Groq(api_key=chat_api)

  def assistente_quimica_fisica(user_prompt:str)->str:
    try:
      if user_prompt.strip()=="15":
        return "Encerrando o chatbot"

      completion=client.chat.completions.create(
          model="llama-3.1-8b-instant",
          messages=[
              {"role":"system","content":"Você é um assistente de química e física"},
              {"role":"user","content":user_prompt},
          ],
          temperature=0,
          max_tokens=1024,
          top_p=1,
          stream=False
      )
      return completion.choices[0].message.content
    except Exception as e:
      print("\nErro no assistente de quimica e fisica")
      return f"Erro: {e}"

  iface = gr.Interface(
      fn = assistente_quimica_fisica,
      inputs = gr.components.Textbox(lines=10, label="Faça uma pergunta sobre quimica ou fisica para nosso assistente"),
      outputs = gr.Textbox(lines=10, label="Resposta"),
      title="Assistente de química e física",
      description="Perguntas sobre Quimíca e Fisíca",
      live=False
  )

  if __name__ == "__main__":
    iface.launch(debug=True, show_error=True)
```
