```python
from google.colab import files
import os
os.makedirs("/contents/documentos", exist_ok=True)
print('Pasta criada em /content/documentos')
```

```python
#
from llama_index.core import SimpleDirectoryReader
documentos = SimpleDirectoryReader(input_dir='/content/documentos')
```

```python
docs = documentos.load_data()
print(f'Quantidade de documentos carregados {len(docs)}')
```

```python
print(docs[0].get_metadata_str())
```

```python
from llama_index.core.node_parser import SentenceSplitter
node_parser = SentenceSplitter(chunk_size=1200)


node = node_parser.get_nodes_from_documents(docs, show_progress=True)
print(f'Quantidade de nodes gerados {len(node)}')
```

```python
from llama_index.llms.google_genai import GoogleGenAI
from llama_index.embeddings.google_genai import GoogleGenAIEmbedding
from llama_index.core import Settings


llm = GoogleGenAI(
    model='gemini-2.5-flash', # Changed to a valid LLM model, assuming 'gemini-2.5-flash' was a typo or non-existent
    api_key=api
)

embed_model = GoogleGenAIEmbedding(
    model='gemini-embeddings-001',
    api_key=api
)

Settings.llm = llm
Settings.embed_model = embed_model
print('LLM e embeddings configurados')
```

```python
from llama_index.core import VectorStoreIndex
index = VectorStoreIndex(node,embed_model=embed_model)
print('Index criada com sucesso')
```

```python
from llama_index.core.base.embeddings.base import similarity

query_engine = index.as_query_engine(llm=llm,similarity_top_k=1)
response = query_engine.query("Quais grãos estão disponiveis")
print(response)
```
